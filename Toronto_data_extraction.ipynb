```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
table_mn = pd.read_html(url)
```

```python
len(table_mn)
```

```python
new_table = table_mn[0]
```

```python
new_table.shape
```

```python
new_table.head()
```

```python
df = pd.DataFrame(columns = ['test'])
df
```

```python
number_of_col = new_table.shape[1]
for i in range(number_of_col):
    new_table.rename(columns={i: 'test'},inplace = True)
    df1=pd.DataFrame(new_table['test'])
    df = df.append(df1)
    new_table.rename(columns={'test': i}, inplace = True)
df
```

### we got the dataframe that need to separate PostalCode, Borough and Neighborhood

```python
df=df.reset_index(drop=True)
df.tail()
```

```python
df=df.assign(PostalCode = df['test'].str[:3], borough_neigh = df['test'].str[3:])
df.head(3)
```

```python
#take the part of borough_neigh before '(' that will Borough column 
# to avoid error add column xyz and then drop
df[['Borough', 'Neighborhood', 'xyz']] = df.borough_neigh.apply(lambda x: pd.Series(str(x).split("(")))
```

```python
df.head(3)
```

```python
# taking required column
newdf = df[['PostalCode','Borough','Neighborhood']]
newdf
```

```python
# removing ')' and replacing '/' by ','
newdf['Neighborhood'] = newdf['Neighborhood'].str.replace(')',"")
newdf['Neighborhood'] = newdf['Neighborhood'].str.replace('/',",")
newdf.sample(5)
```

```python
# eliminating 'Not Assigned rows'
newdf =newdf[newdf['Borough']!='Not assigned']
newdf.reset_index(drop = True)
```

```python
print("The number of rows in the cleaned dataset is {}.".format(newdf.shape[0]))
```
